```python
import re
import pandas as pd
import datetime as dt
```

```python
f = open("WhatsApp Chat with SNC CMSA SEM 6.txt",'r',encoding='utf-8')
data = f.read() #--In the form of a string, the content of the file has been read
print(data)
```

```python
pattern = r'\s-\s'
```

```python
message = [re.split(pattern, line)[1] if re.split(pattern, line)[1:] else line for line in data.split('\n')]
message
# len(message)
```

```python
date_pattern = r'^(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[apAP][mM]\s-)\s'
dates = [re.findall(date_pattern, line)[0] if re.findall(date_pattern, line) else '' for line in data.split('\n')]
print(dates)
```

```python
df = pd.DataFrame({'user_message': message, 'message_date': dates})
date_formats = ['%m/%d/%y, %I:%M %p -', '%d/%m/%y, %I:%M %p -']
for fmt in date_formats:
    try:
        df['message_date'] = pd.to_datetime(df['message_date'], format=fmt)
    except ValueError:
        pass


df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
#separate username and messages
users = []
messages = []
for message in df['user_message']:
    entry = message.split(':')
    if entry[1:]: #username
        users.append(entry[0])
        messages.append(entry[1])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'],inplace = True)
df.head()

```

```python
# extracting year  and month
df['year'] = df['date'].dt.year
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
df.head()
```

```python
#extracting day,hour,minute
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df.head()
```

```python
import re

# Function to extract links from a message
def extract_links(message):
    # Regular expression pattern to match URLs
    url_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F])|(?:[^\s]))+'
    
    # Find all URLs in the message
    links = re.findall(url_pattern, message)
    
    return links

# Apply the function to extract links from each message in the DataFrame
df['links'] = df['message'].apply(extract_links)
```

```python
#remove group notifications and omitted media
temp = df[df['user'] != 'group_notification']
temp = temp[temp['message'] != r' <Media omitted>']
```

```python
# remove stop words
f = open("stop_hinglish.txt",'r')
stop_words = f.read()
# print(stop_words)
```

```python
words = []
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
```

```python
# To convert the most common words into dataframe
from collections import Counter
pd.DataFrame(Counter(words).most_common(10))
```

```python
df['day_name'] = df['date'].dt.day_name()
df['day_name'].value_counts()
```

```python
df.head()
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour-1))
```

```python
df['period']  = period
df.sample(5)
```

```python
import seaborn as sn
plt.figure(figsize = (20,6))
sn.heatmap(df.pivot_table(index='day_name',columns='period',values='message',aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
